```python
selected_features = df[["popularity", "danceability", "energy", "tempo"]]
selected_features.head()
```

```python
df["duration_min"] = df["duration_ms"] / 60000
df[["duration_ms", "duration_min"]].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[["energy", "danceability"]] = scaler.fit_transform(df[["energy", "danceability"]])
df[["energy", "danceability"]].head()
```

```python
from sklearn.preprocessing import Normalizer

norm = Normalizer()
df[["loudness", "tempo"]] = norm.fit_transform(df[["loudness", "tempo"]])
```

```python
df["explicit"] = df["explicit"].astype(int)
df["explicit"].value_counts()
```

```python
df["popularity"].describe()
```

```python
plt.scatter(df["energy"], df["danceability"])
plt.xlabel("Energy")
plt.ylabel("Danceability")
plt.title("Energy vs Danceability")
plt.show()
```
